```python
import os

# 다운로드 받은 파일을 판다스의 read_csv 를 통해 읽어옵니다.
# 파일을 읽어온 후 shape 로 행과 열의 수를 출력합니다.
# csv_path = '/content/drive/MyDrive/health/data/data_2.csv'
csv_path = '/content/drive/MyDrive/아이펠해커톤_2/data_dis01.csv'
df2 = pd.read_csv(csv_path, encoding='cp949')

df = df2.copy()
pd.set_option('display.max_columns', None) # 컬럼 수가 많으므로 요약되어 표시되지 않게 합니다.
# sample, head, tail : 데이터 미리보기
df # 데이터를 확인합니다.
```

```python
df_f = df.copy()
df_f = df_f.iloc[:,[2,3,4,6,5]]
```

```python
df_f
```

```python
from sklearn.pipeline import make_pipeline 
from sklearn.preprocessing import StandardScaler 
standard_scaler = StandardScaler() 
scaled_df = pd.DataFrame(standard_scaler.fit_transform(df_f.iloc[:,0:4]), columns=df_f.iloc[:,0:4].columns) # scaled된 데이터
scaled_df

```

```python
db_df = scaled_df.copy()
```

```python
db_df
```

```python
db_df_100 = db_df.copy()
db_df_100['DIS'] = df['DIS']
db_df_100
```

```python
db_df_s = db_df_100[db_df_100['DIS'] == 1].sample(n=20000)
for i in range(2,5):
  sample_df = db_df_100[db_df_100['DIS'] == i].sample(n=20000)
  db_df_s = pd.concat([db_df_s,sample_df])
db_df_s
```

```python
db_df_s.reset_index(drop=True,inplace=True)
db_df_s
```

```python
db_df_s = db_df_s.sample(frac=1).reset_index(drop=True)
db_df_s
```

```python
db_df_s_100 = db_df_s.copy()
```

```python
db_df_s = db_df_s.iloc[:,:4]
db_df_s
```

```python
for i in range(1,5):
  print((db_df_s_100['DIS'] == i).sum())
```

```python
from sklearn.cluster import DBSCAN
```

```python
min_samples_list = [5,10,100,500,1000]
eps_list = [0.3,0.5,0.7,1.0,1.5]
```

```python
for i in min_samples_list:
  for j in eps_list:  
    db_model = DBSCAN(eps=j, min_samples=i)

    # 군집화 모델 학습 및 클러스터 예측 결과 반환
    db_model.fit(db_df_s)
    db_df_s_100[f'cluster{i}_{j}'] = db_model.fit_predict(db_df_s)
```

```python
for i in min_samples_list:
  for j in eps_list:
    print(f'{i},{j}',db_df_s_100[f'cluster{i}_{j}'].unique())
```

```python
for i in min_samples_list:
  for j in eps_list:
    print(pd.crosstab(db_df_s_100['DIS'],db_df_s_100[f'cluster{i}_{j}']))
```
