## 데이터 라벨링
- 비만 :  시서스 또는 가르시니아(체중 감량에 큰 영향 없다는 연구결과가 있음)
- 빈혈 : 철분제
- 당뇨 : metformine- 또는 오메가3
- 간 :  우루사

**BMI(비만여부)** + 시서스 or 가르시니아(체중 감량에 큰 영향 없다는 연구결과가 있음)

```python
def class_bmi_div (bmi) :
    if bmi < 18.5 :  # 18.5 미만
        return "정상(비만)" # 저체중
    elif 18.5 <= bmi < 23 :
        return "정상(비만)" # 정상체중
    elif 23 <= bmi < 25 :
        return "시서스" # 과체중
    elif 25 <= bmi < 30 :  # 질환의심: 30 이상
        return "시서스" # 경도비만
    elif 30 <= bmi < 35 :
        return "시서스" # 중정도비만
    elif 35 <= bmi :
        return "시서스" # 고도비만

df['영양제 복용여부']=df['bmi'].apply(class_bmi_div)
```

```python
df.head()
```

---

**혈색소(빈혈여부)** + 철분제

```python
cond1 = (df['성별'] == '1')
cond2 = (df['혈색소'] < 12)
cond3 = (df['혈색소'] < 13)

df.loc[cond1&cond2, '빈혈여부'] = "철분제" # 여자 and 혈색소 < 12 : 빈혈
df.loc[~cond1&~cond2, '빈혈여부'] = "정상(빈혈)" # 여자 and 혈색소 >= 12 : 빈혈 아님
df.loc[~cond1&cond3, '빈혈여부'] = "철분제" # 남자 and 혈색소 < 13 : 빈혈
df.loc[~cond1&~cond3, '빈혈여부'] = "정상(빈혈)" # 남자 and 혈색소 >= 13 : 빈혈 아님 

df['영양제 복용여부']=df['영양제 복용여부'].astype('object')
```

```python
df.head()
```

---

**식전혈당(공복혈당) 당뇨여부** + 메트포르만(metformine-) or 오메가3


```python
def trans(tr) :
    if 70 <= tr < 100 :
        return "정상(당뇨)" # 정상
    elif 100 <= tr < 126 :
        return "메트포르민" # 비정상(공복혈당장애)
                 # 100 ~ 110 : 100 미만 대비 당뇨병 위험도가 3.17배, 110 ~ 126 : 100 미만 대비 당뇨병 위험도가 7.42배
    else :
        return "메트포르민" # 당뇨병
        # 정상(저혈당, 정상) / 비정상(공복혈당장애, 당뇨병)

df['영양제 복용여부']=df['식전혈당(공복혈당)'].apply(trans).astype('object')
```

```python
df.head()
```

---

**AST/ALT비율, GTP/ALT비율(간 수치)** + 우루사

**AST/ALT 비율**

```python
df['AST/ALT비율'] = df['(혈청지오티)AST']/df['(혈청지오티)ALT'] # AST/ALT비율
```

**감마지티피/ALT비율**

```python
df['GTP/ALT비율'] = df['감마 지티피']/df['(혈청지오티)ALT']
```

**알코올성 간질환 여부**

```python
# AST/ALT비율 > 1
# 감마지티피/ALT > 2.5
# AST는 300을 넘지 않음
# ALT는 종종 정상
# AST/ALT비율이 2~3 이상인 경우 알코올성 간질환을 강하게 시사한다.

cond1 = ((df['AST/ALT비율']>1)&(df['GTP/ALT비율']>2.5)&(df['(혈청지오티)AST']<=300))

df.loc[cond1, '영양제 복용여부'] = '우루사'
df.loc[~cond1, '영양제 복용여부'] = '정상(간)'
```
