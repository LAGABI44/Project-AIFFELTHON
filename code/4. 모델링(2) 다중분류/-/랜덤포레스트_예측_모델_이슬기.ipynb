```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/drive/MyDrive/슬기로운 건강생활/data/국민건강보험공단_건강검진정보_20211229.CSV', encoding='cp949')
df
```

```python
# train_a = train.truncate(after='100',axis=0)

df = df.drop(labels=range(100000, 1000000), axis=0)

df
```

```python
df.info()
```

```python
# 1차 가공 : 사용하지 데이터 삭제
df.drop(['기준년도','가입자 일련번호','시도코드','치아우식증유무','치석','데이터 공개일자'], axis=1, inplace=True)

# sample, head, tail을 통해 데이터를 미리보기 합니다.
df.head()
```

```python
# NaN 값이 한개라도 존재하는 행 제거
df.dropna(axis=0, how='any', subset=['성별코드', '연령대 코드(5세단위)', '신장(5Cm단위)',
       '체중(5Kg 단위)', '허리둘레', '시력(좌)', '시력(우)', '청력(좌)', '청력(우)', '수축기 혈압',
       '이완기 혈압', '식전혈당(공복혈당)', '총 콜레스테롤', '트리글리세라이드', 'HDL 콜레스테롤', 'LDL 콜레스테롤',
       '혈색소', '요단백', '혈청크레아티닌', '(혈청지오티)AST', '(혈청지오티)ALT', '감마 지티피', '흡연상태',
       '음주여부', '구강검진 수검여부'], inplace=True)
```

- float을 int로 변환
- 반올림

```python
df = df.round(0).astype(int)
```

```python
df.tail()
```

```python
train = df
```

# 체중(5Kg 단위) 예측

```python
# feature와 target 분리
yt=np.array(train['체중(5Kg 단위)'])
xt=train.drop(['체중(5Kg 단위)'], axis=1)
xt=np.array(xt)

# 데이터 정규화
scaler=MinMaxScaler()
xt=scaler.fit_transform(xt)

# train test 데이터 split
xtrain,xtest,ytrain,ytest=train_test_split(xt, yt, test_size=0.2, random_state=42)
```

```python
# 랜덤포레스트 without tuning
rf = RandomForestClassifier(random_state=17)
rf.fit(xtrain, ytrain)
forest_predictions = rf.predict(xtest)

accuracy_score(ytest, forest_predictions)
forest_predictions = rf.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
# 데이터 100만 개 사용

print(confusion_matrix(ytest, forest_predictions))
```

```python
print(classification_report(ytest, forest_predictions))
```
